```python
import cv2
import os

video_path = '/content/project_video.mp4'  # Path to the downloaded video
output_dir = '/content/video_frames'  # Directory to save extracted frames

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

cap = cv2.VideoCapture(video_path)
frame_count = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_filename = os.path.join(output_dir, f"frame_{frame_count:04d}.jpg")
    cv2.imwrite(frame_filename, frame)  # Save frame as image
    frame_count += 1

cap.release()
print(f"Frames extracted: {frame_count}")
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
import easyocr
from deep_sort_realtime.deepsort_tracker import DeepSort

# Initialize DeepSort with appropriate parameters
deepsort = DeepSort(max_age=30, nn_budget=70, max_iou_distance=0.7)

# Initialize EasyOCR reader and YOLO models
reader = easyocr.Reader(['en'], gpu=False)
coco_model = YOLO('yolov8n.pt')  # Path to pre-trained COCO model for cars
license_plate_detector = YOLO('/content/trained_model_cars.pt')  # Custom license plate model

# Car class ID for COCO dataset
vehicles = [2]

# Function to detect and format license plate text
def read_license_plate(license_plate_crop):
    license_plate_crop_gray = cv2.cvtColor(license_plate_crop, cv2.COLOR_BGR2GRAY)
    _, license_plate_crop_thresh = cv2.threshold(license_plate_crop_gray, 90, 255, cv2.THRESH_BINARY)
    detections = reader.readtext(license_plate_crop_thresh)
    for detection in detections:
        bbox, text, score = detection
        return text, score
    return None, None

# Open video file
input_video_path = '/content/project_video.mp4'
output_video_path = '/content/output_video.mp4'
cap = cv2.VideoCapture(input_video_path)

# Check if the input video opened successfully
if not cap.isOpened():
    print("Error: Could not open input video.")
    exit()

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Set up VideoWriter
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# Process frames
frame_nmr = -1
while cap.isOpened():
    frame_nmr += 1
    ret, frame = cap.read()
    if not ret:
        break

    # Debug: Confirm frame is captured
    print(f"Processing frame {frame_nmr}")

    # Vehicle detection
    detections = coco_model(frame)[0]  # Get the detection results for vehicles
    detections_ = []

    for detection in detections.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = detection
        if int(class_id) in vehicles:
            detections_.append([[x1, y1, x2, y2], score])  # Prepare for DeepSORT

    # Update DeepSORT tracks
    tracks = deepsort.update_tracks(detections_, frame=frame)
    print(f"Tracks in frame {frame_nmr}: {[track.track_id for track in tracks]}")  # Debug track IDs

    # License plate detection
    license_plates = license_plate_detector(frame)[0]

    # Loop through license plates and match them to vehicles
    for license_plate in license_plates.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = license_plate
        for track in tracks:
            track_id = track.track_id
            tx1, ty1, tx2, ty2 = track.to_tlbr()
            if x1 > tx1 and y1 > ty1 and x2 < tx2 and y2 < ty2:
                license_plate_crop = frame[int(y1):int(y2), int(x1):int(x2)]
                license_plate_text, license_plate_text_score = read_license_plate(license_plate_crop)
                if license_plate_text:
                    print(f"Detected license plate: {license_plate_text} in frame {frame_nmr}")

                    # Annotate frame
                    cv2.rectangle(frame, (int(tx1), int(ty1)), (int(tx2), int(ty2)), (0, 255, 0), 2)
                    cv2.putText(frame, f'ID {track_id}', (int(tx1), int(ty1)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
                    cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 0, 255), 2)
                    cv2.putText(frame, f'{license_plate_text}', (int(x1), int(y1)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

    # Write annotated frame to output video
    out.write(frame)

# Release video resources
cap.release()
out.release()

print("Processing complete. Check output video at:", output_video_path)
```
